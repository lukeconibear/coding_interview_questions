```python
def find_fixed_point1(nums): # time O(n), space O(1)
    for index, value in enumerate(nums):
        if index == value:
            return index

def find_fixed_point2(nums): # time O(lgn), space O(lgn), as used binary search tree recursively
    return _find_fixed_point_helper(0, len(nums), nums)

def _find_fixed_point_helper(index_low, index_high, nums):
    if index_low == index_high:
        return None

    index_mid = int((index_low + index_high) / 2)
    if nums[index_mid] == index_mid:
        return index_mid
    if nums[index_mid] < index_mid:
        return find_fixed_point_helper(index_mid + 1, index_high, nums)
    else:
        return find_fixed_point_helper(index_low, index_mid, nums)

def find_fixed_point3(nums): # time O(lgn), space O(n), as used binary search tree iteratively
    index_low = 0
    index_high = len(nums)

    while (index_low != index_high):
        index_mid = int((index_low + index_high) / 2)
        if nums[index_mid] == index_mid:
            return index_mid
        if nums[index_mid] < index_mid:
            index_low = index_mid + 1
        else:
            index_high = index_mid

    return None


# where the index equals the value
print(find_fixed_point1([-5, 1, 3, 4])) 
print(find_fixed_point2([-5, 1, 3, 4]))
print(find_fixed_point3([-5, 1, 3, 4]))
# 1
```
