```python
class Node:
    def __init__(self, value, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right
        
    def __repr__(self):
        result = f'{self.value}'
        if self.right:
            result += f' {self.right}'
        return result
          
class Solution:
    def flatten1(self, head): # time O(n), space O(n)
        in_order_nodes = self.get_nodes_in_order(head, [])
        for index in range(0, len(in_order_nodes) - 1):
            left_node = in_order_nodes[index]
            right_node = in_order_nodes[index + 1]
            left_node.right = right_node
            right_node.left = left_node

        return in_order_nodes[0]

    def get_nodes_in_order(self, head, array):
        if head is not None:
            self.get_nodes_in_order(head.left, array)
            array.append(head)
            self.get_nodes_in_order(head.right, array)

        return array

    def flatten2(self, head): # time O(n), space O(h)
        left_most, _ = self.flatten_tree(head)
        return left_most

    def flatten_tree(self, node):
        if node.left is None:
            left_most = node
        else:
            left_subtree_left_most, left_subtree_right_most = self.flatten_tree(node.left)
            self.connect_nodes(left_subtree_right_most, node)
            left_most = left_subtree_left_most

        if node.right is None:
            right_most = node
        else:
            right_subtree_left_most, right_subtree_right_most = self.flatten_tree(node.right)
            self.connect_nodes(node, right_subtree_left_most)
            right_most = right_subtree_right_most

        return [left_most, right_most]

    def connect_nodes(self, left, right):
        left.right = right
        right.left = left
        
    def test(self):
        head = Node(1)
        head.left = Node(2)
        head.right = Node(3)
        head.left.left = Node(4)
        head.left.right = Node(5)
        head.right.left = Node(6)
        head.left.right.left = Node(7)
        head.left.right.right = Node(8)

        result = Node(4)
        result.right = Node(2, left=result)
        result.right.right = Node(7, left=result.right)
        result.right.right.right = Node(5, left=result.right.right)
        result.right.right.right.right = Node(8, left=result.right.right.right)
        result.right.right.right.right.right = Node(1, left=result.right.right.right.right)
        result.right.right.right.right.right.right = Node(6, left=result.right.right.right.right.right)
        result.right.right.right.right.right.right.right = Node(3, left=result.right.right.right.right.right.right)
        
        print(Solution().flatten1(head))
        #print(Solution().flatten2(head))
        print(result)
        
        
print(Solution().test())
```
