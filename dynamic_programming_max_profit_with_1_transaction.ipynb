```python
def max_profit_with_1_transactions1(array): # time O(n2), space O(1)
    max_profit = 0
    for i in range(len(array)):
        for j in range(i, len(array)):
            max_profit = max(max_profit, array[j] - array[i])
            
    return max_profit


def max_profit_with_1_transactions2(array): # time O(n), space O(1)
    max_current_price = 0
    max_profit = 0
    for price in array[::-1]:
        max_current_price = max(max_current_price, price)
        max_profit = max(max_profit, max_current_price - price)

    return max_profit


print(max_profit_with_1_transactions1([9, 11, 8, 5, 7, 10]))
print(max_profit_with_1_transactions2([9, 11, 8, 5, 7, 10]))
```
